# NSD train fMRI voxel mean / std

```python
import os
import numpy as np
import torch

DATA_PATH = "/mnt/shared-storage-user/ai4sdata2-share/maiweijian/BrainVL/data/"
SUBS = (1, 2, 5, 7)
```

```python
for sub in SUBS:
    npy_path = os.path.join(DATA_PATH, f"nsd/subj0{sub}/nsd_train_fmri_scale_sub{sub}.npy")
    all_fmri = np.load(npy_path).astype(np.float32)
    all_fmri = all_fmri.mean(axis=1)
    all_fmri = torch.tensor(all_fmri, dtype=torch.float32).squeeze(1)

    voxel_mean = all_fmri.mean(dim=0)  # [D]
    voxel_std = all_fmri.std(dim=0)  # [D]

    out_path = os.path.join(
        DATA_PATH, f"nsd/subj0{sub}/nsd_train_fmri_voxel_stats_sub{sub}.pt"
    )
    torch.save({"voxel_mean": voxel_mean, "voxel_std": voxel_std}, out_path)
    print(f"sub={sub}  saved  {out_path}")
    print(f"  voxel_mean shape={tuple(voxel_mean.shape)}  voxel_std shape={tuple(voxel_std.shape)}")
```
